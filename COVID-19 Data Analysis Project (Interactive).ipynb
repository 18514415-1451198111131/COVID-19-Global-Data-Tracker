```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
import plotly.express as px
import geopandas as gpd
```

```python
#Loading the dataset
df = pd.read_csv('full_grouped.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Inspection of the dataset structure
df.columns     # Checking column names
df.head()      # Previewing the first few rows
df.info()      # General info
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Data Cleaning
num_cols = ['Recovered', 'Active', 'New cases', 'New deaths', 'New recovered']
df[num_cols] = df[num_cols].fillna(0)
df = df.dropna(subset=['Confirmed', 'Deaths', 'Date'])
```

```python
#Interactivity: User Inputs
country_dropdown = widgets.Dropdown(options=df['Country/Region'].unique(),value='Kenya',description='Country:')
```

```python
min_date = df['Date'].min()
max_date = df['Date'].max()
```

```python
date_range = widgets.SelectionRangeSlider(
    options=[d.strftime('%Y-%m-%d') for d in pd.date_range(min_date, max_date)],
    index=(0, len(pd.date_range(min_date, max_date))-1),description='Date Range:',layout={'width': '800px'})
display(country_dropdown, date_range)
```

```python
# Filtering Function
def filter_data(country, start_date, end_date):
    mask = ((df['Country/Region'] == country) &(df['Date'] >= pd.to_datetime(start_date)) &(df['Date'] <= pd.to_datetime(end_date)))
    return df[mask]
```

```python
# Run the filter using widget values
filtered_df = filter_data(country_dropdown.value, date_range.value[0], date_range.value[1])
```

```python
# 📈 Total Confirmed Cases Plot
plt.figure(figsize=(10,6))
plt.plot(filtered_df['Date'], filtered_df['Confirmed'], label='Confirmed Cases')
plt.title(f'Total Confirmed COVID-19 Cases in {country_dropdown.value}')
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 📉 Death Rate Plot
plt.figure(figsize=(10,6))
filtered_df['Death Rate'] = filtered_df['Deaths'] / filtered_df['Confirmed']
plt.plot(filtered_df['Date'], filtered_df['Death Rate'], label='Death Rate')
plt.title(f'COVID-19 Death Rate in {country_dropdown.value}')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 📊 Bar Chart for Latest Date's New Cases
latest_date = filtered_df['Date'].max()
latest_data = df[df['Date'] == latest_date]
plt.figure(figsize=(10,6))
sns.barplot(data=latest_data, x='Country/Region', y='New cases')
plt.title(f'New COVID-19 Cases by Country on {latest_date.date()}')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
